```python
#import library NetworkX untuk membuat graf
import networkx as nx

#import matplotlib untuk menggambar graf
import matplotlib.pyplot as plt

#Graph Eropa
#Membuat graf kosong dengan tipe undirected graph
eropa = nx.Graph()

#Membuat daftar hubungan antar kota Eropa
edges_eropa = [
    ("Oradea", "Zerind", 71),
    ("Oradea", "Sibiu", 151),
    ("Zerind", "Arad", 75),
    ("Arad", "Sibiu", 140),
    ("Arad", "Timisoara", 118),
    ("Timisoara", "Lugoj", 111),
    ("Lugoj", "Mehadia", 70),
    ("Mehadia", "Drobeta", 75),
    ("Drobeta", "Craiova", 120),
    ("Craiova", "Sibiu", 146),
    ("Craiova", "Pitesti", 138),
    ("Sibiu", "Fagaras", 99),
    ("Sibiu", "Rimnicu Vilcea", 80),
    ("Rimnicu Vilcea", "Pitesti", 97),
    ("Pitesti", "Bucharest", 101),
    ("Bucharest", "Giurgiu", 90),
    ("Bucharest", "Urziceni", 85),
    ("Urziceni", "Hirsova", 98),
    ("Urziceni", "Vaslui", 142),
    ("Hirsova", "Eforie", 86),
    ("Vaslui", "Iasi", 92),
    ("Iasi", "Neamt", 87)
]
#Memasukkan edge dan bobot ke graf Eropa
eropa.add_weighted_edges_from(edges_eropa)

#Graph Indonesia
#Membuat graf kosong Jawa
indonesia = nx.Graph()

#Membuat daftar hubungan antar kota Jawa
edges_indonesia = [
    ("Jakarta", "Bandung", 270),
    ("Jakarta", "Cirebon", 327),
    ("Bandung", "Cirebon", 120),
    ("Bandung", "Yogyakarta", 373),
    ("Cirebon", "Semarang", 305),
    ("Cirebon", "Yogyakarta", 210),
    ("Semarang", "Surakarta", 97),
    ("Semarang", "Surabaya", 380),
    ("Surakarta", "Yogyakarta", 60),
    ("Surakarta", "Malang", 370),
    ("Surabaya", "Malang", 94)
]
#Memasukkan edge dan bobot ke graf Indonesia
indonesia.add_weighted_edges_from(edges_indonesia)

#Fungsi menggambar graph
def show_graph(G, pos=None, title=''):

    #Membuat ukuran gambar
    plt.figure(figsize=(12, 8))

    #Jika posisi kosong maka membuat posisi otomatis
    if pos is None:
        pos = nx.spring_layout(
            G,
            seed=42
        )
    #Menggambar node dan edge graph
    nx.draw(
        G,
        pos,
        with_labels=True,
        node_color='lightblue',
        node_size=2200,
        font_color='black',
        font_weight='bold',
        width=2,
        edge_color='gray',
        font_size=10
    )
    edge_labels = nx.get_edge_attributes(
        G,
        'weight'
    )
    nx.draw_networkx_edge_labels(
        G,
        pos,
        edge_labels=edge_labels,
        font_color='darkred',
        font_weight='bold',
        font_size=10
    )
    plt.title(
        title,
        fontsize=16,
        fontweight='bold'
    )

    plt.axis('off')
    plt.show()

#Fungsi pencarian shortest path
def cari_shortest_path(graf):

    #Mengambil semua simpul/kota dari graf
    kota = list(graf.nodes())

    #Perulangan memilih kota pertama
    for i in range(len(kota)):

        #Perulangan memilih kota kedua
        for j in range(i + 1, len(kota)):

            #Menentukan kota asal
            asal = kota[i]

            #Menentukan kota tujuan
            tujuan = kota[j]

            #Mencari jalur terpendek
            path = nx.shortest_path(
                graf,
                source=asal,
                target=tujuan,
                weight="weight"
            )
            #Menghitung total jarak
            jarak = nx.shortest_path_length(
                graf,
                source=asal,
                target=tujuan,
                weight="weight"
            )
            print(asal,"->",tujuan)
            print("Shortest Path :",path)
            print("Total Jarak :",jarak)
            print("-"*50)

#Menampilkan graph Eropa
show_graph(
    eropa,
    title="Graph Kota Eropa"
)

#Menampilkan graph Indonesia
show_graph(
    indonesia,
    title="Graph Kota Jawa"
)

#Menampilkan hasil shortest path Eropa
print("SHORTEST PATH GRAF EROPA")
cari_shortest_path(eropa)

#Menampilkan hasil shortest path Indonesia
print("\nSHORTEST PATH GRAF INDONESIA")
cari_shortest_path(indonesia)
```
